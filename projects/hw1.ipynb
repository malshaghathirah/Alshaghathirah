```python
print("Treatment column unique values:", df['treatment'].unique())
print("Donation outcome:", df['gave'].unique())
```

```python
# Calculate donation rate by treatment group
summary = df.groupby("treatment")["gave"].mean().reset_index()
summary.columns = ["Treatment Group", "Donation Rate"]
summary
```

```python
import statsmodels.api as sm

# X = treatment dummy, Y = donation outcome
X = df[["treatment"]]
X = sm.add_constant(X)
y = df["gave"]

model = sm.OLS(y, X).fit()
print(model.summary())
```
